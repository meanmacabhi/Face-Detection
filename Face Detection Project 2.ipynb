```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('C:\\Users\\Abhirup Dey\\Desktop\\AIML\\Apr 24 ML Day 18 Project\\Projects\\PRJ Face Detection\\Data Set\\people2.jpg')

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load the pre-trained Haar Cascade classifiers for face and eye detection
face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Detect faces with adjusted parameters
faces = face_detector.detectMultiScale(
    gray_image,
    scaleFactor=1.1,  # Lowered to detect smaller faces
    minNeighbors=4,   # Lowered to detect more faces
    minSize=(30, 30)  # Kept as default
)

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    roi_gray = gray_image[y:y + h, x:x + w]
    roi_color = image[y:y + h, x:x + w]
    
    # Detect eyes within the face ROI with adjusted parameters
    eyes = eye_detector.detectMultiScale(
        roi_gray,
        scaleFactor=1.05,  # Lowered to detect smaller eyes
        minNeighbors=5,    # Lowered to detect more eyes
        minSize=(15, 15),  # Decreased to detect smaller eyes
        maxSize=(60, 60)   # Increased to detect larger eyes
    )
    
    # Draw rectangles around detected eyes
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 0, 255), 2)

# Convert the image from BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the resulting image
plt.figure(figsize=(10, 5))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
